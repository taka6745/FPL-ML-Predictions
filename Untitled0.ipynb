```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the shapefiles
queensland_shapefile = gpd.read_file("/content/Restricted_areas.shp")


# Plotting
fig, ax = plt.subplots(1, figsize=(10, 10))

# Plot the Queensland state polygon
queensland_shapefile.boundary.plot(ax=ax, linewidth=1, color="blue", label="Queensland")


# Legend
plt.legend()

# Show the plot
plt.show()
```
